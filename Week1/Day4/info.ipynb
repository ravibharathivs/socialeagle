# Playwright

1. Browser context
2. pages (browser tabs)
3. Css,XPath(selectors), text context used to select the area on the browser

## Installing Playwright

1. install using pip
    * pip install playwright
    * playwright install (for specific browser for playwright, this needs isolated browser)
