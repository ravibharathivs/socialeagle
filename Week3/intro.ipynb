# Langchain

1. What is Langchain?
    *   Langchain is a python framework that helps you build apps powered by AI framework models (like chatgpt) by connecting them with tools like:
        *  Documents
        *  Search
        *  APIs
        *  Memory
        *  Agents
    *   You can think of LangChain as: Language model + tools = Smart AI app

2. Why do weneed Langchain?
    * Wihtout langchain:
      *  you must handle prompts, memory, and logic manually
      *  you write long code for each step
      *  Hard to connect LLMs to tools like databases,files or web search
    * With Langchain:
      *  Makes buildding LLM apps easier and faster
      *  Gives you ready - made chains, agents, memory, retrivers
      *  Lets you create complex flows using simple blocks
    

3. Why Not just use other tools like transformers, OpenAI API, or GPT - 4 directly?

    |Feature         | Raw OpenAI/Transformer         | Langchain          |
    |:----------------|:--------------------------------|:--------------------|
    |Simple prompt   | yes  | yes |
    | prompt templates| no | yes |
    |Memory/Chat history| Manual | Built-in|
    |Document search (RAG) | Difficult | Easy with loaders retrivers |
    |Agents / Tools | build from scratch | Pre-built agents |
    |Chain / Flow logic| Manuall Scripting | Ready chains|

    LangChain = Fast track to LLM - powered apps

    
    

4. What can you build LangChain?
   *    Chatbots with memory
   *    Question answering apps using PDFs or websites
   *    AI agents that can use tools (calculator,browser)
   *    AutoGPT-style apps
   *    Voice-based assistants (with plugins)
