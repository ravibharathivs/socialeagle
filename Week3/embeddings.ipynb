## Embeddings

converting a text or word or sentences into numbers (vector values) called embeddings. so that AI can compare them and can understand it. in simple term GPS uses coordinates to find places, embeddings use vectors to find similar meanings.

### Real life analogy
 Think of each sentences as a point on a map. sentences with similar meanings will be closer together on this map even if th words are different. 

### Why use embeddings?
For semantic search (find ansers based on meaning, not exact words).

To connect chunks of documents with user questions. used in RAG to build Q&A over PDF, chat with docs. etc


Embedding ===========> Turning text into numbers (vectors) that hold meanings.
Embedding model ============> Tool that creates those vectors (e.g OpenAI, HF)

```python
from langchain_openai import OpenAIEmbeddings
openai_embed = OpenAIEmbeddings(openai_api_key="{your_key}",model="text-embedding-3-large")
text_embed = "This is the sample text of embeddings"
embedd = openai_embed.embed_query(text_embed)
print(embedd)
```

```python
from langchain_huggingface import HuggingFaceEmbeddings
embedd2 = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
```
