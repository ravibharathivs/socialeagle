# Retriever

when we ask question to model it will return the correct answers, these relevent ans fetched using this retriever.

*   A retriever is a tool that fetches the most relevant document chunks from a vector store based on your question.
-   Vector store = Memory
-   Retriever = Brains search tool
-   LLM = a person who giving a answers

so when a question is asked, the retriever finds the best matching pieces of info, then passes them to the LLM to answer.


## Retriever Langchain workflow:
-   pdf/Text ---> chunk --> Embed --> store in vector DB
-   Ask question --> Retriever gets relevant chunks --> LLM answer using those chunks

```python
retriever = vectorstore.as_retriever()
```

```python
query = "langchain?"

# For this LangChain version, use the internal method with a run_manager argument.
answer = retriever._get_relevant_documents(query, run_manager=None)

print(answer)
```
