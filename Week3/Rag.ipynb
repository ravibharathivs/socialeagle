## RAG (Retrieval Augmented Generation)

### What is RAG?
simple explanation of RAG: combining search + generation
Real-world analogy: Like a student using a book + own wordds to answer.

### Why we use RAG?
Limitations of standalone LLMs (they forget facts,hallucinate)
Need for external knowledge.


### How RAG is different from chatGPT?
ChatGPT (pure generation)
RAG (Retrieval + generation) it reads from doc and generate the ansers from it.

### How Retrieval Works
Keyword search vs semantic search
Introduction to embeddings (basic idea : turning into numbers)

### Components of RAG

-   Retriever => Finds the relevant documents from a knowledge base

-   Generator => users the reieved data to anser or create a content

-   Knowledge  store - database or document collection (PDFs,text,webpages)

### Flow of RAG

*   user question -> retriever -> generator -> final answer

### Use cases of RAG

-   Homework helper using text books PDFs
-   personalized assistant with your own notes
-   chatbot for personal documents

## Types of RAG

1.  Self RAG
2.  Corective RAG
3.  Fusion RAG
4.  Advance RAG
5.  Speculative RAG

### 1.Self RAG

*   Self RAG means the system checks its own answer and, if unsure, retrieves more information and tries again. Its like a student writing an answer, double-checking it with notes, and improve it

### 2.Corrective RAG

*   Corrective RAG identifies the generated answer is wrong or it misleading the answer. if its not correct it corrects by fetching better sources or rewording the answer. Like peer-review for RAG.

### 3.Fusion RAG (combine multiple docs)

*   Fusion RAG retrives multiple documents and merges their information to generate a comprehensive,non-redundant answer. Useful when no single doc has the full picture.

### 4.Advanced RAG

*   Advanced RAG uses external toools, multi-step reasoning, or planning agents (e,g LangChain Agents) to complex queries.
*   It can retrieve, process and combine information in stages.

*   flow ====>    guess -> retrieve -> reason -> correct

### 5.Speculative RAG

*   Instead of retrieveing first, Speculative RAG starts by guessing an answer based on the question. then it retrieves documents to refine or verify it. Helps when initial retrieval is weak.

*   Flow ====> guessed answer ===> vector DB ====> docuemts fetched ===> LLM  ===> answer
