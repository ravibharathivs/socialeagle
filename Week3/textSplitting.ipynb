## Text Splitting

### 1. What is text splitting?
Text splitting means breaking a large document into smaller chunks, so the AI model can read it piece by piece.

### 2. Why do we need slit text?
    LLMs (like GPT) can only handle a limited number of words at a time (Called tokens)

#### Splitting help us:   
    Avoid errors like "context too long"
    Feed meaningfull chunks to the model
    Use retrieval-based Q&A

#### How text splitting works (Simple Analogy)
    Imagine a big book. you can't memorize the whole book in one go right?
    so we need to read 1 page at a time (that spliting the page for books is called chunks)



*   Chunk_size        ===============> Max characters per chunk
*   chunk_overlap     ===============> Characters repeated between chunks

## Coding

```python
pip install -U langchain-text-splitters
```

```python
from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import CharacterTextSplitter

test1 = PyPDFLoader("pvc_price tag.pdf")
loded_text = test1.load()

fulltext = "\n".join([te.page_content for te in loded_text])

text_splitter = CharacterTextSplitter(chunk_size=100, chunk_overlap=0)
texts = text_splitter.split_text(fulltext)
```

```python
texts
```

## Recursive text splitter

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter


text_splitter = RecursiveCharacterTextSplitter(chunk_size=100, chunk_overlap=0)
recursivetext = text_splitter.split_text(fulltext)
```
