# Knowledge Graph RAG

Traditional rag will fetches the content only matched with input keywords, but this knowledge graph fetch and validate the fetched content with its history,availability resources. so the hallucination of this type rag will be reduced up to 99%.
 

### How a documents are converted into Knowledge Graph RAG

|Traditional Rag | KG - RAG|
|----------------|---------|
|Document load| Document load|
|Text split|NER (named entity recognition) + Entity Linking |
|Embeddings (openai) | Relation Extraction (Arrow)|
|Vector DB | Temporial tagging (timing noted. validate with previous history)|


ex:
if we asked question about a person, this KG-RAG will not say only current status of the person, it searches and fetches all historical and future plans for that person. which means it not only fetch the content from the document which we feeded it also search from external tools if anything configured.
