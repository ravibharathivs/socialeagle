## Data ingestion (Data loader)

1. what is Data ingestion?
 *  Data ingestion = loading external info (like from a book or website) into memory so that the ai can unserstand and anser questions based on it.

### Flow
 Data ingestion --------> Text Splitter -----> Embeddings -----> Vector store  -------> Retriver
 

## Examples of source of loading data

* Websites
* research papers
* from disk
* from any external sources
